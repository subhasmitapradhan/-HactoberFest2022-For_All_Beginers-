```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
# Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

train_generator = train_datagen.flow_from_directory(
    '/content/dataset/train',  # Path to training data directory
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical')

test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    '/content/dataset/test',  # Path to testing data directory
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical')
```

```python
path = '../content/dataset/train/'
```

```python
class_names = sorted(os.listdir(path))
num_classes = len(class_names)

img_size = (128, 128, 3)

print('classes: ', class_names)
```

```python
import cv2

labels = []
images = []

print('images:\n')
for cl in class_names:
    print(cl, end=' -> ')
    for img in os.listdir(path + cl):
        label = np.zeros(num_classes)
        label[class_names.index(cl)] = 1
        labels.append(label)

        image = np.asarray(cv2.resize(cv2.imread(path + cl + '/' + img, cv2.IMREAD_COLOR), img_size[0:2])[:, :, ::-1])#That line remove Blue

        #image = np.asarray(cv2.resize(cv2.imread(data_path + cl + '/' + img, cv2.COLOR_BGR2RGB), img_size[0:2]))
        images.append(image)
    print('done')


labels = np.asarray(labels)
images = np.asarray(images)

print(f'\n\nlabels shape: {labels.shape}')
print(f'images shape: {images.shape}')
```

```python
from random import randint

fig, axs = plt.subplots(3, 3, figsize=(16, 16))#Original was 10,6

for x in range(3):
    for y in range(3):
        i = randint(0, len(images))

        axs[x][y].imshow(images[i])


        axs[x][y].set_xticks([])
        axs[x][y].set_yticks([])
        axs[x][y].set_xlabel(class_names[np.argmax(labels[i])])


plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Reduce dataset size by selecting a subset
subset_size = 1500  # Adjust this number to your desired subset size
random_indices = np.random.choice(images.shape[0], subset_size, replace=False)
images_subset = images[random_indices]
labels_subset = labels[random_indices]

# Split the reduced dataset into training and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(images_subset, labels_subset, test_size=0.10, random_state=42)

print(f'train images shape: {X_train.shape}\ntrain labels shape: {Y_train.shape}\n')
print(f'validation images shape: {X_val.shape}\nvalidation labels shape: {Y_val.shape}\n')
```

```python
from tensorflow.keras.applications import DenseNet121, EfficientNetB0

# Define the base models
densenet_base = DenseNet121(weights='imagenet', include_top=False, input_shape=(192, 192, 3))
efficientnet_base = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(192, 192, 3))

# Freeze all layers in the pre-trained models
for layer in densenet_base.layers:
    layer.trainable = False

for layer in efficientnet_base.layers:
    layer.trainable = False

# Concatenate the features from both models
densenet_features = densenet_base.output
efficientnet_features = efficientnet_base.output
merged_features = Concatenate()([densenet_features, efficientnet_features])

# Add a dense layer for classification
flattened_features = Flatten()(merged_features)
dense_layer = Dense(256, activation='relu')(flattened_features)
output_layer = Dense(train_generator.num_classes, activation='softmax')(dense_layer)

# Create the hybrid model
hybrid_model = Model(inputs=[densenet_base.input, efficientnet_base.input], outputs=output_layer)

# Compile the model
hybrid_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
